```python
# pdf2image 모듈 import
from pdf2image import convert_from_path

# 이미지 파일로 변환할 pdf 파일 이름 변수 선언
f = "/content/drive/MyDrive/text/1. dataset/우리은행/예적금/1. 상품설명서_「우리 첫거래우대 정기예금」.pdf"

# 이미지 파일을 저장할 디렉토리명 선언
s = "/content/drive/MyDrive/text/1-1. dataset to image/"
# pdf 파일에서 img 객체 리스트 반환해서 pages 변수로 받기
pages = convert_from_path(f)

# 이미지 파일 저장 - 페이지 순으로 이미지파일명1.jpg, 이미지파일명2.jpg...
for i, page in enumerate(pages):
    image_filename = f"{s}1. 상품설명서_「우리 첫거래우대 정기예금」{i+1}.jpg"
    page.save(image_filename, 'JPEG')
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# 이미지 파일을 저장하고 있는 디렉토리 경로
directory_path = "/content/drive/MyDrive/text/1-1. dataset to image/"

# 디렉토리 내의 모든 파일 목록을 불러옴
file_list = os.listdir(directory_path)

# 파일 목록을 순회하면서 JPEG 파일만 불러와서 출력
for file_name in file_list:
    if file_name.endswith('.jpg'):  # JPEG 파일 확인
        image_path = os.path.join(directory_path, file_name)
        img = Image.open(image_path)
        plt.imshow(img)
        plt.title(file_name)  # 이미지 위에 파일 이름을 제목으로 표시
        plt.axis('off')
        plt.show()
```

```python
from PIL import Image
import pytesseract

# 이미지 불러오기
image_path = '/content/drive/MyDrive/text/1-1. dataset to image/1. 상품설명서_「우리 첫거래우대 정기예금」1.jpg'
img = Image.open(image_path)

# OCR 실행
text = pytesseract.image_to_string(img, lang='kor')
print("Extracted Text:", text)

# 이미지 출력
plt.figure(figsize=(10, 10))  # 이미지 크기 설정
plt.imshow(img)
plt.axis('off')  # 축 정보 끄기
plt.show()
```
